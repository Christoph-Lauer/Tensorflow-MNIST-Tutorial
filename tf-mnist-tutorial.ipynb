# Tensorflow-MNIST Tutorial

This is a very basic example for the usage of tensorflow with the MNIST dataset. The MNIST Dataset is a large database of handwritten digits where near human performace can be archived using a committee of neuronal networks.



The tensorflow implementation is very simple.

```python
# import von TF
import tensorflow as tf
```

```python
# Handgeschriebene Ziffern laden
mnist = tf.keras.datasets.mnist
```

```python
# Aufteilung in Training- und Testset
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
# TF Bilderkennungsmodell
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(512, activation=tf.nn.relu),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
# Crossentropy fr die 10 Zahlen Klassen
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Modellfitting 
model.fit(x_train, y_train, epochs=5)
```

```python
# Evaluation
model.evaluate(x_test, y_test)
```
